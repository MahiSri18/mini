```python
import pandas as pd 

import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import os 
```

```python
plt.style.use('Solarize_Light2')
```

```python
cric_df = pd.read_csv('C:/Users/user/Downloads/archive/Men ODI Team Batting Stats.csv')
```

```python
cric_df
```

```python
cric_df.set_index('Country', inplace= True)
```

```python
cric_df['Matches Won'].plot(kind ='bar',figsize=(15,10),color=('red','black'))
plt.xlabel('Countries',size=14)
plt.ylabel('No. of matches',size=14)
                                                
plt.title('MATCHES WON BY COUNTRIES',size= 22, color = 'brown')
plt.show()
```

```python
cric_df['Matches Lost'].plot(kind='bar',figsize=(15,10),color=('brown','violet'))
plt.title('MATCHES LOST BY COUNTRIES',size=22,color='red')
plt.xlabel('Countries',size=14)
plt.ylabel('No. of matches',size=14)
plt.show()
```

```python
cric_df.sort_values(by='Team Matches Played',ascending= False,inplace= True)
```

```python
cric_df
```

```python
cric_df['Avg Runs Per Six Balls Batting'].transpose().plot(kind='line',stacked=False, figsize=(15,7),color='green')
plt.title('avg runs per over countires')
plt.ylabel('avg runs',size=24)
plt.xlabel('countries',size=24)
plt.show()
```

```python
cric_df['Win/Loss Ratio'].plot(kind= 'pie', figsize=(15,8),
                            autopct='%.2f%%',
                               labels=None,
                            startangle=90,
                               pctdistance=0.8,
                            shadow=True)
plt.title('Ratio of teams winning/losing',size=18)
plt.ylabel('Win/Loss Ratio',size=18)
plt.legend(labels=cric_df.index,loc='upper right')
plt.axis('equal')

plt.show()
```

```python
cric_df['Avg Runs Per Wicket Batting'].plot(kind='area',figsize=(12,10),color='grey')
plt.title('Area Graph of runs per Wicket batting',size=22,color='orange')
plt.show()
```

```python
cric_df.sort_values(by='Win/Loss Ratio',ascending=False, inplace=True)
```

```python
cols_to_drops = ['Number Of Team Innings Batting','Highest Team Score Batting','Lowest Completed Score Batting']
cric_df.drop(cols_to_drops, axis=1, inplace=True)
```

```python
teams_com = graphs_cric
fig=plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)


teams_com['Avg Runs Per Wicket Batting'].transpose().plot(kind='line',ax=ax1,figsize=(15,10))
ax1.set_title('line plot of Avg Runs Per Wicket Batting')



teams_com['Avg Runs Per Six Balls Batting'].transpose().plot(kind='line',ax=ax2,figsize=(15,10))
ax2.set_title('line plot of Avg Runs Per Six Balls Batting')

plt.show()
```

```python
teams_com = graphs_cric
fig=plt.figure()
ax1 = fig.add_subplot(2,1,2)
ax2 = fig.add_subplot(2,1,1)


teams_com.transpose().plot(kind='area',ax=ax1,figsize=(15,10))
ax1.set_title('Area plot of Stats of top 5 countries',size=16)


teams_com.transpose().plot(kind='line',ax=ax2,figsize=(15,15))
ax2.set_title('Line plot  of Stats of top 5 countries',size=16)

plt.show()
```
